```python
import pandas as pd

BASE_PATH = "/content/drive/MyDrive/Customer-Retention-Analysis"

df = pd.read_csv(
    f"{BASE_PATH}/data/processed/clean_merged_data.csv"
)
```


Monthly Orders

```python
df['order_purchase_timestamp'] = pd.to_datetime(
    df['order_purchase_timestamp']
)

df['month'] = df[
    'order_purchase_timestamp'
].dt.to_period('M')

monthly_orders = df.groupby('month')['order_id'].nunique()

monthly_orders.head()
```

Calculates total unique orders placed each month

```python
monthly_orders = df.groupby(
    'month'
)['order_id'].nunique()

print(monthly_orders.head())
```

Calculates total revenue generated each month

```python
monthly_revenue = df.groupby(
    'month'
)['payment_value'].sum()

print(monthly_revenue.head())
```

Calculates number of orders placed by each customer

```python
customer_orders = df.groupby(
    'customer_unique_id'
)['order_id'].nunique()

print(customer_orders.head())
```
