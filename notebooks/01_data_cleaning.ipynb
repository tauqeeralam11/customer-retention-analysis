```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers = pd.read_csv(
    f"{BASE_PATH}/data/raw/olist_customers_dataset.csv"
)

orders = pd.read_csv(
    f"{BASE_PATH}/data/raw/olist_orders_dataset.csv"
)

payments = pd.read_csv(
    f"{BASE_PATH}/data/raw/olist_order_payments_dataset.csv"
)
```

```python
data cleaning
A. Convert Dates
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(
    orders['order_purchase_timestamp']
)
```

```python
B. Remove Duplicates
```

```python
orders.drop_duplicates(inplace=True)
```

```python
C. Keep Delivered Orders Only
```

```python
orders = orders[orders['order_status'] == 'delivered']
```

```python
Merge Tables
```

```python
merged_df = orders.merge(
    customers,
    on='customer_id',
    how='left'
)

merged_df = merged_df.merge(
    payments,
    on='order_id',
    how='left'
)
```

```python
Save Processed Data
```

```python
merged_df.to_csv(
    f"{BASE_PATH}/data/processed/clean_merged_data.csv",
    index=False
)
```
